```python
import matplotlib.pyplot as plt
import numpy
import pandas as pd

dict_places = {'all cards':[2.323215, 1.840958], 'yellow cards':[2.247114, 1.784523], 'red cards':[0.076101, 0.056434]}

frame_d = pd.DataFrame(dict_places)

fig, ax = plt.subplots()
fig.set_figwidth(18.7)

ax.set_title('Toutes cartes à étranger en domicile')

ax.pie(frame_d['all cards'].values, shadow=True)
ax.legend(['en domicile', 'à étranger'], bbox_to_anchor =(0.15, 0.25))
plt.show()
```
